## Parse Messenger Messages

```python
# ---- Reset converted html to txt file
htmlfile = "data/raw/shuvom_messenger.html"
txtfile = "data/raw/shuvom_messenger.txt"
```

```python
# ---- Just getting all raw sentences said by me and recepient
f = open(txtfile, "r", encoding="utf16")
me = open(toguyfile, "a", encoding="utf16")
guy = open(fromguyfile, "a", encoding="utf16")
myname = "David Inho Lee\n"
hisname = "Shuvom Sadhuka\n"

# ---- Skip first five lines
for i in range(5):
    f.readline()
    
one = f.readline()
two = f.readline()
three = f.readline()
# ---- Read lines and parse them until the end
while (not("Say hi to your new Facebook friend" in two)):
    # If nothing was said, shift down two sections
    if ((myname == three) or (hisname == three) or (myname == two[1:]) or (hisname == two[1:])):
        one = three
        two = f.readline()
        three = f.readline()
    elif (myname == one):
        # Append the text to words I say
        # Exclude links
        if (not("https://" in two) and not("Click for video:\n" == two)):
            me.write(two)
        one = f.readline()
        two = f.readline()
        three = f.readline()
    elif (hisname == one):
        # Append the text to words he says
        # Exclude links
        if (not("https://" in two) and not("Click for video:\n" == two)):
            guy.write(two)
        one = f.readline()
        two = f.readline()
        three = f.readline()
    else:
        # If not start of the text, shift down a section
        one = two
        two = three
        three = f.readline()

f.close()
me.close()
guy.close()
```

## Parse Text Messages

```python
# ---- Parse Guys Messages

# ---- Just getting all raw sentences said by me and recepient
txtfile = ["data/raw/albert_text.txt", "data/raw/max_text.txt", "data/raw/nicholas_text.txt"]
hisname = ["Albert", "Max", "Nicholas"]
me = open(toguyfile, "a", encoding="utf16")
guy = open(fromguyfile, "a", encoding="utf16")

for i in range(3):
    f = open(txtfile[i], "r", encoding="utf8")
    him = hisname[i]
    
    one = f.readline()
    two = f.readline()
    three = f.readline()
    # ---- Read lines and parse them until the end
    while ((two != "\n") and (two != "")):
        if ("Me" in one):
            # Append the text to words I say
            if (not("https://" in two)):
                me.write(two)
            if (three != "\n"):
                # If there is more text, then continue reading
                two = three
                three = f.readline()
            else:
                one = f.readline()
                two = f.readline()
                three = f.readline()
        elif (him in one):
            # Append the text to words they say
            if (not("https://" in two)):
                guy.write(two)
            if (three != "\n"):
                # If there is more text, then continue reading
                two = three
                three = f.readline()
            else:
                one = f.readline()
                two = f.readline()
                three = f.readline()
        else:
            # Should not reach
            print("Should not be here!")

f.close()
me.close()
guy.close()
```
